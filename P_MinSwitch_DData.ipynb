```python
##Version 3 To minimize the switching for single runtime dynamic data\
# P_MinSwitch_DData: Project, Minimization of Switching, Dyanamic Runtime Data.
Using -1 for success and -100 for failure.


import numpy as np; np.random.seed(0)
import random
import seaborn as sns; sns.set_theme()
import math
import time


for apple in range(5):
    
    
    dt=[[]*10]*10
    avail_per=7
    for i in range(10):
        l = random.sample([0]*avail_per + [1]*(10-avail_per), 10)
        dt[i]=l

    data = np.transpose(dt)
    q_table = [[0 for i in range(10)] for j in range(10)] 
    q_table2 = [[0 for i in range(10)] for j in range(10)] 

     
    sumr=[0]*10
    ch_lst=[]
    # To generate the data Used for Switching
    def Min_Switch():

        sum_s=[]


        avail_cal=[0]
        for rows in range(10):
            rowsum=0
            for cols in range(10):
                   rowsum+=q_table2[rows][cols]
            sumr[rows]+=rowsum    

        for i in range(10):
            sum_s.append(sumr[i])

        sum_s.sort()

        for datas in range(10):
            for datar in range(10):

                if(sum_s[datas]==sumr[datar]):
                    ch_lst.append(datar+1)


    #Learning From Prediction
    def Q_Learner(p_Schedule):
        hit=0
        miss=0
        Miss=[0]*10
        Miss_C=[-1]*10
        Hit=[0]*10
        Hit_C=[0]*10
        switch=0 


        for i in range(10):

            if i>0:
                if p_Schedule[i]!=p_Schedule[i-1]:
                    switch+=1
            if data[p_Schedule[i]-1][i]==0: # if predicted channel is available
                hit+=1
                Hit[i]=1
                Hit_C[i]=p_Schedule[i]
            else:                       #if predicted channel is Missing
                miss+=1
                Miss[i]=1
                Miss_C[i]=p_Schedule[i] # Missing Channel list

    #     print(Miss)

        print(f"Miss Rate: {miss*10} % || Hit rate: {hit*10}")

        for i in range(10):
            if(Miss[i]==1):   
#                 print(f"Miss at T{i+1}, Repredict:")
                switch+=2
                q_table[Miss_C[i]-1][i]= (q_table[Miss_C[i]-1][i]+1) 
                q_table2[Miss_C[i]-1][i]= (q_table2[Miss_C[i]-1][i]+100)
            elif(Hit[i]==1):
                q_table2[Hit_C[i]-1][i]= (q_table2[Hit_C[i]-1][i]+1)


        print("Switching is :",switch)      
        print(Miss_C)
                          # Returing Predicted channel found not available 


    #Prediction for Exploration to Learn     
    def Predictor():
        prediction=[0]*10       # Prediction Array initialized 0
        for i in range(10):
            q_column=[sub[i] for sub in q_table2]    # Taking i th coloumn of Q table
            min_ele=min(q_column)           #Finding the Minimum entry in the column
            min_ele_count=q_column.count(min_ele)  # Number of times the min_ele Occured


            if min_ele_count==1:      #element occured only once
                channel_index=q_column.index(min_ele)
                prediction[i]= channel_index+1 # Channel index starts from 1
            else:
                min_ele_index=[0]*min_ele_count    # To store indexes of min elemens
                j=0     # loop variable
                for i1 in range(10):
                    if q_column[i1]==min_ele and j!=min_ele_count:
                        min_ele_index[j]=i1
                        j+=1
                channel_index=random.choice(min_ele_index)  

    #             print("???????   ",min_ele_index)
                prediction[i]=channel_index+1     
    #     Print_data(q_table,1)
        print("\n\n")
#         print("=================================================")
#         print(f"Min ele count is: {min_ele_count}")
        return prediction


    # Prediction to Minimize the switching
    def Predictor2():
        prediction=[0]*10
        for it3 in range(10):
            prediction[it3]=ch_lst[0]
        return prediction    


    #Data Printing
    def Print_data(arr,mode):
        if mode==1:   # To print in matrix mode
            for row in arr:
                print(' '.join([str(arr) for arr in row]))
        if mode==2:     # To print in graphical representation
            uniform_data = arr
            ax = sns.heatmap(uniform_data,linewidths=.5, cmap="Dark2", cbar=False)    

    print("New data is: ")
    Print_data(data,1)
    
    def Train_Agent2():
       
        explr=50
        explt=10
        switch=2
        
        print("\n\n=============== Exploration=================")
        while explr!=0:
            explr-=1
            pre_Schedule=[random.randint(1,10) for i in range(10)]
            Q_Learner(pre_Schedule)
        
        
        print("\n\n=============== Exploitation=================")
        while explt!=0:
            
            explt-=1
            pre_Schedule=Predictor()
            Q_Learner(pre_Schedule)
            print("Prediction is Exploitation:",pre_Schedule)
        
        
        print("\n\n============Minimizing the Switches=====")
        while switch!=0: 
            switch-=1
            Min_Switch()
            print(sumr,"\n")
            pre_Schedule=Predictor2()
            Q_Learner(pre_Schedule)
            print("Prediction to minimize Switching: ",pre_Schedule)

            Print_data(q_table2,1)    
            Print_data(data,2)



    ##############################
    Train_Agent2()        
    print("\n\n")



```
